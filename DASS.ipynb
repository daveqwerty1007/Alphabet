```python
# 完成时间过短或者过长的无效数据,按照3sigma原理筛选
def three_sigma(data):
    mean = np.mean(data)
    std = np.std(data)
    down = mean-3*std
    up = mean+3*std
    
    outlier = []
    for i in range(data.shape[0]):
        if(data.iloc[i]<down)|(data.iloc[i]>up):
            outlier.append(i)
        else:
            continue
    return outlier
```

标记label

```python
def depression_scores(list):
    scores = 0
    k = [3, 5, 10, 13, 16, 17, 21, 24, 26, 31, 34, 37, 38, 42]
    for i in range(len(k)):
        a = (k[i]-1)*3
        scores += int(list[a])
    return scores
```

```python
def anxiety_scores(list):
    scores = 0
    k = [2, 4, 7, 9, 15, 19, 20, 23, 25, 28, 30, 36, 40, 41]
    for i in range(len(k)):
        a = (k[i]-1)*3
        scores += int(list[a])
    return scores
```

```python
def stress_scores(list):
    scores = 0
    k = [1, 6, 8, 11, 12, 14, 18, 22, 27, 29, 32, 33, 35, 39]
    for i in range(len(k)):
        a = (k[i]-1)*3
        scores += int(list[a])
    return scores
```

```python
def d_mark(n):
    if n>=0 and d<=9:
        level = 0
    elif n>=10 and n<=13:
        level = 1
    elif n>=14 and n<=20:
        level = 2
    elif n>=21 and n<=27:
        level = 3
    elif n>=28:
        level = 4
    return level
```

```python
def a_mark(n):
    if n>=0 and d<=7 :
        level = 0
    elif n>=8 and n<=9:
        level = 1
    elif n>=10 and n<=14:
        level = 2
    elif n>=15 and n<=19:
        level = 3
    else:
        level = 4
    return level
```

```python
def s_mark(n):
    if n>=0 and d<=14:
        level = 0
    elif n>=15 and n<=18:
        level = 1
    elif n>=19 and n<=25:
        level = 2
    elif n>=26 and n<=33:
        level = 3
    else:
        level = 4
    return level
```

```python
# depression scores (3, 5, 10, 13, 16, 17, 21, 24, 26, 31, 34, 37, 38, 42)
# anxiety scores (2, 4, 7, 9, 15, 19, 20, 23, 25, 28, 30, 36, 40, 41)
# stress scores (1, 6, 8, 11, 12, 14, 18, 22, 27, 29, 32, 33, 35, 39)
sum_d =[]
sum_a = []
sum_s = []
Total =[]
for i in range(df2.shape[0]):
    d = depression_scores(df2.iloc[i]) 
    a = anxiety_scores(df2.iloc[i])
    s = stress_scores(df2.iloc[i])
    t = round((d_mark(d)+a_mark(a)+s_mark(s))*11/13)
    sum_d.append(d)
    sum_a.append(a)
    sum_s.append(s)
    Total.append(t)
df2['Depression scores'] = sum_d
df2['Anxiety scores'] = sum_a
df2['Stress scores'] = sum_s
df2['Total'] = Total
df2
```
